```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Documents\Desktop\project 1\orignal heart  data(2).csv") 
df.head()
```

```python
df.info()


print("\n‚ùì Missing Values:")
print(df.isnull().sum())

print("\nüìà Summary Statistics:")
display(df.describe())
```

```python
# Show missing values again just to confirm
print("Missing values before handling:")
print(df.isnull().sum())

# Option 1: Fill numeric columns with median
# (you can also use mean, depending on what works best)
df.fillna(df.median(numeric_only=True), inplace=True)

# Optionally drop rows with any remaining missing values
df.dropna(inplace=True)

# Double check
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Preview cleaned data
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Step 3: Normalize numeric features
scaler = MinMaxScaler()

# List of numeric columns to scale
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Apply MinMaxScaler
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Check the scaled columns
print(" Normalized numerical features:")
df[numerical_cols].head()
```

```python
print(" Columns in your current DataFrame:")
print(df.columns.tolist())
```

```python
# Step 5: Save the cleaned dataset
df.to_csv("cleaned_data.csv", index=False)

print(" Cleaned data saved successfully as 'cleaned_data.csv'")
```
